```python
import numpy as np
from skopt import Optimizer
import matplotlib.pyplot as plt
import copy
import math
from scipy.optimize import curve_fit
```

```python
# Set Range
x_max = 4.0
x_min = 1.0
x_gap = x_max-x_min
y_max = 40.0
y_min = 10.0
y_gap = y_max-y_min
z_max = 100.0
```

```python
# Data collected by BO process(ratio, time, plqy)

rx = (np.array([2.4,1.0,4.0,1.0,4.0,2.4,2.4,2.5,2.1,1.9,2.2,2.8,2.2,2.2,2.1,1.8])-x_min)/x_gap
ry = (np.array([21.0,10.0,10.0,40.0,40.0,30.0,10.0,10.0,13.0,40.0,10.0,40.0,40.0,20.0,21.0,21.0])-y_min)/y_gap
rz = np.array([55.25,35.54,19.512,44.892,20.68,50.99,52.69,50.64,55.02,45.98,52.03,30.12,46.32,56.16,55.88,48.54])/z_max
```

### Get Data with Bayesian Optimization
using Gaussian Process


```python
bounds = [(0.0, 1.0),(0.0, 1.0)]
epoch = 100
result = np.array([[0.0,0.0] for i in range(epoch)])
for k in range(epoch):
    if(k%10==0):
        print("epoch : "+str(k))
    opt_r = Optimizer(bounds, base_estimator="gp", n_initial_points = 0)
    for i in range(0,len(rx)):
        opt_r.tell([rx[i],ry[i]],-rz[i])
    next_rx = opt_r.ask(n_points = 1)
    for i in next_rx:
        result[k][0]=i[0]*x_gap+x_min
        result[k][1]=i[1]*y_gap+y_min
```

```python
# Next recommendation of optimal condition 
print(result)
```
